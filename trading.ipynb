```python
def capital_growth(rate, base, count=40, additional=0, reinvest= .8, min_aside = 0, max_aside = 0):
    mnth_rate = rate/12 / 100
    for i in range(count):
        compound = earn = base * mnth_rate
        set_aside = 0

        if min_aside and max_aside:
            if earn < min_aside / 2:
                set_aside = earn / 2
            else:
                set_aside = min(earn/2, max_aside)
            compound = earn - set_aside
    
        elif reinvest != 1:
            compount = earn * reinvest
            set_aside = earn - compound
        base = base + compound + additional
        print('iteration: %10d  earned: %12.2f  base: %12.2f  kept: %8.f' % (i+1, earn, base, set_aside))

capital_growth(390, 9000, additional=0, min_aside=0, max_aside=0)
```

```python
from cwutils import dicts as dd
def get_dict(items, key_fn):
    res = defaultdict(list)

    for d in items:
        res[key_fn(d)].append(d)
    return res

by_label = lambda items: get_dict(items, lambda item: item.get('label'))
by_month = lambda items: get_dict(items, lambda item: item['date'][:2])

per_month = by_month(dicts)
per_month = {k:by_label(v) for k,v in per_month.items()}
sum_items = lambda items: sum([float(i['amount']) for i in items])
sums_per_month = {k:{lk: sum_items(lv) for lk,lv in v.items()} for k,v in per_month.items()}
april_sum = sums_per_month['04']
#per_month['04']['crypto']
#per_month['04']
def monthly(month):
    return per_month[month]
def month_category(month, category):
    return monthly(month).get(category)

def without_fluff(monthly_labeled, *to_remove):
    rm_keys = to_remove or ('food', 'books', 'booze', 'schwab', 'crypto', 'cms', 'income', 'one_time', 'video')
    return  dd.without_keys(monthly_labeled, *rm_keys)
def month_sum(month):
    data = without_fluff(sums_per_month[month])
    return data
    
print(sum([v for v in without_fluff(april_sum).values()]))
no_fluff = {k:without_fluff(v) for k,v in per_month.items()}
no_fluff['04']
print(month_sum('04'))
sum_values = lambda labeled_sums: sum([v for v in labeled_sums.values()])
sum_values(month_sum('04'))
month_category('03', 'supplements')
```
