```python
def dict_print(d, tabs = 0):
    for k,v in d.items():
        if isinstance(v, dict):
            print(tabs * '\t', '%s:' % k)
            dict_print(v, tabs+1)
        else:
            print(tabs * '\t', '%s:' % k, v)

dict_print(p1.positions())
```
